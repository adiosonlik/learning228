```python
cat_train = ['cc_2', 'cc_4', 'cc_6']
```

```python
lst=[]
for i in cat_train:
    for j in train[i]:
        lst.append(j)
lst
```

```python
le = LabelEncoder()

le.fit(lst)

for col in cat_train:
    train[col] = le.transform(train[col])

train
```
